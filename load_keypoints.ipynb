```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
# loads from saved file
with open("pose_data.jsonl") as f:
    data = [json.loads(line) for line in f if line.strip()]
```

```python
video_index = 0
frame_index = 1
bodypart_index = 1
len(data)
```

```python
def show_pose_animation(video_id, interval=50, padding=400, figsize=(6,6)):
    num_frames = len(data[video_id]['keypoints'])

    fig, ax = plt.subplots(figsize=figsize)
    scatter = ax.scatter([], [], c='red', s=40)

    # Compute axis limits
    all_keypoints = []
    for frame in data[video_id]['keypoints']:
        for kp in frame:
            all_keypoints.append(kp[:2])
    all_keypoints = np.array(all_keypoints)
    x_all, y_all = all_keypoints[:,0], all_keypoints[:,1]
    ax.set_xlim(np.min(x_all)-padding, np.max(x_all)+padding)
    ax.set_ylim(np.max(y_all)+padding, np.min(y_all)-padding)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")

    def update(frame_id):
        frame_data = data[video_id]['keypoints'][frame_id]
        keypoints = np.array([kp[:2] for kp in frame_data])
        scatter.set_offsets(keypoints)
        ax.set_title(f"YOLOv11 Pose â€” Video {video_id}, Frame {frame_id}")
        return scatter,

    ani = FuncAnimation(
        fig,
        update,
        frames=num_frames,
        interval=interval,
        blit=False
    )

    # Prevent the static figure from displaying
    plt.close(fig)

    return HTML(ani.to_jshtml())
```

```python
show_pose_animation(1)  # Show video 50
```

```python
show_pose_animation(40)  # Show video 50
```
